```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-5,5,25)
```

```python
Y = 5*X**3.0 + 5.0*X**2.0 - 12.0*X -2
```

```python
plt.scatter(X,Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.tight_layout()
```

```python
modelo1 = np.poly1d(np.polyfit(X, Y, 1))
modelo2 = np.poly1d(np.polyfit(X, Y, 2))
modelo3 = np.poly1d(np.polyfit(X, Y, 3))
modelo4 = np.poly1d(np.polyfit(X, Y, 4))
modelo5 = np.poly1d(np.polyfit(X, Y, 5))
```

```python
y1 = modelo1(X)
y2 = modelo2(X)
y3 = modelo3(X)
y4 = modelo4(X)
y5 = modelo5(X)
```

```python
plt.scatter(X,Y)
plt.plot(X,y1,color='purple',label='Grau 1')
#plt.plot(X,y2,color='red',label='Grau 2')
#plt.plot(X,y3,color='green',label='Grau 3')
#plt.plot(X,y4,color='orange',label='Grau 4')
#plt.plot(X,y5,color='cyan',label='Grau 5')
plt.xlabel('X')
plt.ylabel('Y')
#plt.legend()
plt.tight_layout()
```
